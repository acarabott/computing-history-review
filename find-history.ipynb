```python
import os
import subprocess
import glob
from bs4 import BeautifulSoup

# depends on pdfgrep https://pdfgrep.org/
# brew install pdfgrep on os x
```

```python
RESULTS_DIR = 'results'
```

```python
files = []
file_types = ['txt', 'html', 'pdf']
for root, dirs, files in os.walk('./data'):
    for file_type in file_types:
        for filename in glob.iglob('./**/*.{}'.format(file_type), recursive=True):
            files.append(filename)
```

```python
def check_courses(patterns, out_path):
    with open(os.path.join(RESULTS_DIR,out_path), 'w') as out_file:
        for i, f in enumerate(files):
            print('checking {}/{} {} '.format(i, len(files), f))
            ext = os.path.splitext(f)[1]
            cmd = 'pdfgrep' if ext == '.pdf' else 'grep'

            for pattern in patterns:
                try:
                    result = subprocess.check_output([cmd, '-i', pattern, f])
                    out_file.write('{}\n'.format(f))
                    cleantext = BeautifulSoup(result, 'html.parser').text
                    out_file.write('\t{}\n'.format(cleantext))
                except subprocess.CalledProcessError as e:
                    pass
```

```python
check_courses(['history'], 'history.txt')
```

```python
patterns = [
    'turing machine',
]
check_courses(patterns, 'turing-machine.txt')
```

```python
patterns = [
    'big data',
]
check_courses(patterns, 'big-data.txt')
```

```python
patterns = [
    'algorithm',
]
check_courses(patterns, 'algorithm.txt')
```
