```python
import os
import json
import requests
import re
```

```python
folder = '05-cmu'
filename = 'cmu.json'
# folder = '10-princeton'
```

```python
in_path = os.path.join(folder, filename)
in_file = open(in_path, 'r')
j = json.load(in_file)
in_file.close()
```

```python
classes = j.get('classes')
for i, c in enumerate(classes):
    title = re.sub('[^a-z]', '-', c['title'].lower())
    print('{}/{} {}'.format(i, len(classes), title))    
    r = requests.get(c['url'])
    out_path = os.path.join(folder, '{}.html'.format(title))
    with open(out_path, 'w') as out_file:
         out_file.write(r.text)
```
